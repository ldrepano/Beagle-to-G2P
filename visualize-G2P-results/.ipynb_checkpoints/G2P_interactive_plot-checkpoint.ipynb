```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import stats
import plotly.io as pio
import plotly.graph_objs as go
```

```python
gene='PPM1D'
editor='CBE'
```

```python
# Open the G2P output file
g2p_output = pd.read_csv("G2P_PPM1D_O15297_protein_features_CBE.csv", delimiter=',') 
g2p_output
```

```python
g2p_output
```

```python
# Remove Nan values from z-scores
g2p_output = g2p_output.dropna(subset=['Strongest Z-score'])
g2p_output = g2p_output.dropna(subset=['Mean Z-Score'])
g2p_output
```

```python
# Create data frames for small features
#g2p_output_Binding = g2p_output.dropna(subset=['Binding site (UniProt)'])

g2p_output_Mutagenesis = g2p_output.dropna(subset=['Mutagenesis (UniProt)'])

g2p_output_Phosphorylation = g2p_output.dropna(subset=['Phosphorylation'])

g2p_output_Ubiquitination = g2p_output.dropna(subset=['Ubiquitination'])

#g2p_output_Methylation = g2p_output.dropna(subset=['Methylation'])
```

```python
g2p_output_ClinVar = g2p_output[["residueId","AA","Strongest Z-score","Strongest Z-Score Alt AA","Strongest Z-Score sgRNA","Strongest Z-Score mutation","Missense (ClinVar)", "Synonymous (ClinVar)", "Nonsense (ClinVar)"]]#"Nonsense (ClinVar)"
g2p_output_ClinVar = g2p_output_ClinVar.dropna(subset=["Missense (ClinVar)", "Synonymous (ClinVar)", "Nonsense (ClinVar)" ], how='all') #'Synonymous (ClinVar)', 'Nonsense (ClinVar)'
g2p_output_ClinVar 
```

```python
g2p_output_ClinVar['Missense (ClinVar)'] = g2p_output_ClinVar['Missense (ClinVar)'].str.split(';')

# Explode the list-like values in 'ColumnA' into separate rows
g2p_output_ClinVar = g2p_output_ClinVar.explode('Missense (ClinVar)')
g2p_output_ClinVar
```

```python
# Parse the protein mutation in the way that initial amino acid, its position and the final mutant amino acid are separated in three different columns
g2p_output_ClinVar['BE edit'] = g2p_output_ClinVar['Strongest Z-Score Alt AA'].str[-3:]
g2p_output_ClinVar
```

```python
# Convert columns to string type if they contain non-string data
g2p_output_ClinVar['Missense (ClinVar)'] = g2p_output_ClinVar['Missense (ClinVar)'].astype(str)
g2p_output_ClinVar['Synonymous (ClinVar)'] = g2p_output_ClinVar['Synonymous (ClinVar)'].astype(str)
g2p_output_ClinVar['Nonsense (ClinVar)'] = g2p_output_ClinVar['Nonsense (ClinVar)'].astype(str)

# Now apply string methods
g2p_output_ClinVar['Missense mut AA'] = g2p_output_ClinVar['Missense (ClinVar)'].str[-3:]
g2p_output_ClinVar['Synonymous mut AA'] = g2p_output_ClinVar['Synonymous (ClinVar)'].str[2:5]
g2p_output_ClinVar['Nonsense mut AA'] = g2p_output_ClinVar['Nonsense (ClinVar)'].str[-3:]
```

```python
# See if the base editor mutation matches to the ClinVar mutation
g2p_output_ClinVar['Missense mut AA match'] = g2p_output_ClinVar['BE edit'] == g2p_output_ClinVar['Missense mut AA']
g2p_output_ClinVar['Synonymous mut AA match'] = g2p_output_ClinVar['BE edit'] == g2p_output_ClinVar['Synonymous mut AA']
g2p_output_ClinVar['Nonsense mut AA match'] = g2p_output_ClinVar['BE edit'] == g2p_output_ClinVar['Nonsense mut AA']
g2p_output_ClinVar
```

```python
g2p_output_ClinVar['Missense mut AA match'] = g2p_output_ClinVar['BE edit'] == g2p_output_ClinVar['Missense mut AA']
g2p_output_ClinVar['Synonymous mut AA match'] = g2p_output_ClinVar['BE edit'] == g2p_output_ClinVar['Synonymous mut AA']
g2p_output_ClinVar['Nonsense mut AA match'] = g2p_output_ClinVar['BE edit'] == g2p_output_ClinVar['Nonsense mut AA']
```

```python
counts1 = g2p_output_ClinVar['Missense mut AA'].count()
counts2 = (g2p_output_ClinVar['Missense mut AA match'] == True).sum()

counts3 = (g2p_output_ClinVar['Synonymous mut AA match'] == True).sum()

counts4 = (g2p_output_ClinVar['Nonsense mut AA match'] == True).sum()

print(counts1)
print(counts2)
print(counts3)
print(counts4)
```

```python
g2p_output_ClinVar_Missense = g2p_output_ClinVar[g2p_output_ClinVar['Missense mut AA match'] == True]
g2p_output_ClinVar_Synonymous = g2p_output_ClinVar[g2p_output_ClinVar['Synonymous mut AA match'] == True]
g2p_output_ClinVar_Nonsense = g2p_output_ClinVar[g2p_output_ClinVar['Nonsense mut AA match'] == True]

g2p_output_ClinVar_Nonsense
```

```python
import plotly.graph_objs as go

# Your existing data
x1 = g2p_output['residueId']
y1 = g2p_output['Strongest Z-score']
sgRNA1 = g2p_output['Strongest Z-Score sgRNA']
AA1 = g2p_output['Strongest Z-Score Alt AA']
mutation1 = g2p_output['Strongest Z-Score mutation']

x2 = g2p_output_Phosphorylation['residueId']
y2 = g2p_output_Phosphorylation['Strongest Z-score']
sgRNA2 = g2p_output_Phosphorylation['Strongest Z-Score sgRNA']
AA2 = g2p_output_Phosphorylation['Strongest Z-Score Alt AA']
mutation2 = g2p_output_Phosphorylation['Strongest Z-Score mutation']

#x3 = g2p_output_Acetylation['residueId']
#y3 = g2p_output_Acetylation['Strongest Z-score']
#sgRNA3 = g2p_output_Acetylation['Strongest Z-Score sgRNA']
#AA3 = g2p_output_Acetylation['Strongest Z-Score Alt AA']
#mutation3 = g2p_output_Acetylation['Strongest Z-Score mutation']

x4 = g2p_output_Ubiquitination['residueId']
y4 = g2p_output_Ubiquitination['Strongest Z-score']
sgRNA4 = g2p_output_Ubiquitination['Strongest Z-Score sgRNA']
AA4 = g2p_output_Ubiquitination['Strongest Z-Score Alt AA']
mutation4 = g2p_output_Ubiquitination['Strongest Z-Score mutation']

#x6 = g2p_output_Binding['residueId']
#y6 = g2p_output_Binding['Strongest Z-score']
#sgRNA6 = g2p_output_Binding['Strongest Z-Score sgRNA']
#AA6 = g2p_output_Binding['Strongest Z-Score Alt AA']
#mutation6 = g2p_output_Binding['Strongest Z-Score mutation']

x7 = g2p_output_ClinVar_Missense['residueId']
y7 = g2p_output_ClinVar_Missense['Strongest Z-score']
sgRNA7 = g2p_output_ClinVar_Missense['Strongest Z-Score sgRNA']
AA7 = g2p_output_ClinVar_Missense['Strongest Z-Score Alt AA']
mutation7 = g2p_output_ClinVar_Missense['Strongest Z-Score mutation']

x8 = g2p_output_ClinVar_Synonymous['residueId']
y8 = g2p_output_ClinVar_Synonymous['Strongest Z-score']
sgRNA8 = g2p_output_ClinVar_Synonymous['Strongest Z-Score sgRNA']
AA8 = g2p_output_ClinVar_Synonymous['Strongest Z-Score Alt AA']
mutation8 = g2p_output_ClinVar_Synonymous['Strongest Z-Score mutation']

x9 = g2p_output_ClinVar_Nonsense['residueId']
y9 = g2p_output_ClinVar_Nonsense['Strongest Z-score']
sgRNA9 = g2p_output_ClinVar_Nonsense['Strongest Z-Score sgRNA']
AA9 = g2p_output_ClinVar_Nonsense['Strongest Z-Score Alt AA']
mutation9 = g2p_output_ClinVar_Nonsense['Strongest Z-Score mutation']

x10 = g2p_output_Mutagenesis['residueId']
y10 = g2p_output_Mutagenesis['Strongest Z-score']
sgRNA10 = g2p_output_Mutagenesis['Strongest Z-Score sgRNA']
AA10 = g2p_output_Mutagenesis['Strongest Z-Score Alt AA']
mutation10 = g2p_output_Mutagenesis['Strongest Z-Score mutation']

# Define colors for each category
color_dict = {
    'Strongest Z-Score sgRNA': '#DDDDDD',
    'Phosphorylation(PhosphoSite)': '#1f77b4', # Blue
    'Acetylation(PhosphoSite)': '#ff7f0e',     # Orange
    'Ubiquitination(PhosphoSite)': '#2ca02c',  # Green
    'Methylation(PhosphoSite)': '#8c564b',     # Brown
    'Binding(Uniprot)': '#00BFFF',             # Blue
    'Missense (ClinVar)': '#FF77FF',            # Magenta
    'Synonymous(ClinVar)': '#bcbd22',          # Olive
    'Nonsense(ClinVar)': '#d62728',            # Red
    'Mutagenesis(Uniprot)': '#ffd343'         # Yellow
}

# Define scatter plot traces for each dataset
traces = [
    go.Scatter(x=x1, y=y1, mode='markers', name='Color codes', marker=dict(color='black', size=0))
]

# Add scatter plot traces for other datasets
for i, (x, y, sgRNA, AA, mutation, label) in enumerate([
    (x1, y1, sgRNA1, AA1, mutation1, 'Strongest Z-Score sgRNA'),
    (x2, y2, sgRNA2, AA2, mutation2, 'Phosphorylation(PhosphoSite)'),
    #(x3, y3, sgRNA3, AA3, mutation3, 'Acetylation(PhosphoSite)'),
    (x4, y4, sgRNA4, AA4, mutation4, 'Ubiquitination(PhosphoSite)'),
    #(x6, y6, sgRNA6, AA6, mutation6, 'Binding(Uniprot)'),
    (x7, y7, sgRNA7, AA7, mutation7, 'Missense (ClinVar)'),
    (x8, y8, sgRNA8, AA8, mutation8, 'Synonymous(ClinVar)'),
    (x9, y9, sgRNA9, AA9, mutation9, 'Nonsense(ClinVar)'),
    (x10, y10, sgRNA10, AA10, mutation10, 'Mutagenesis(Uniprot)')
]):
    # Determine marker color based on label
    marker_color = color_dict[label]
    
    # Set marker symbol as square if mutation contains 'Nonsense'
    marker_symbol = ['circle' if 'Nonsense' not in mut and 'Silent' not in mut else 'square' if 'Nonsense' in mut else 'triangle-up' if 'Silent' in mut else 'circle' for mut in mutation]
    
    trace = go.Scatter(x=x, y=y, mode='markers', name=label, marker=dict(size=20, color=marker_color, symbol=marker_symbol, opacity=0.9))
    trace.hoverinfo = 'text'
    trace.text = [f"{label}: residueId={x_val}, Strongest Z-score={y_val}, sgRNA={sgRNA_val}, Alt AA={AA_val}, Mutation={mutation_val}"
                  for x_val, y_val, sgRNA_val, AA_val, mutation_val in zip(x, y, sgRNA, AA, mutation)]
    traces.append(trace)

# Create layout
layout = go.Layout(
    title=f'{gene}',
    xaxis=dict(title='AA position', zeroline=True, zerolinewidth=1, zerolinecolor='black', color='black', showgrid=True, gridcolor='lightgray'),
    yaxis=dict(title='z-score', zeroline=True, zerolinewidth=1, zerolinecolor='black', color='black', showgrid=True, gridcolor='lightgray'),
    legend=dict(x=1.05, y=1, bgcolor='white', borderwidth=2, bordercolor='black'),
    width=1600,  # Set width of the plot
    height=600,  # Set height of the plot,
    plot_bgcolor='white'  # Set background color to white
)

# Create figure
fig = go.Figure(data=traces, layout=layout)

# Show plot
fig.show()
```

```python
pio.write_html(fig, file=f"{gene}_G2P_interactive_{editor}.html")
```
